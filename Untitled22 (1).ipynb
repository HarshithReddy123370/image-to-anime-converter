```python
#import labreries
import cv2  # used for camera and photo
import numpy as np # mathematical and scientific task
import matplotlib.pyplot as plt # visual representaion and graphical  and flow cahrts
from google.colab import files # it is used to uplod  amd imp
from PIL import Image # Import the Image class from PIL
#from IPython.display import Image  #Added for Colab
```

```python
# Upload an image
uploaded = files.upload()
```

```python

# Read the uploaded image
for filename in uploaded.keys():
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB(red,blue , green)
```

```python

    # Convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
    # Apply median blur
    gray_blur = cv2.medianBlur(gray, 5)
```

```python
    # Detect edges using adaptive thresholding
    edges = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                  cv2.THRESH_BINARY, 9, 9)
```

```python
    # Apply bilateral filter to smoothen colors
    color = cv2.bilateralFilter(img, 9, 300, 300)
```

```python
    # Combine edges with the filtered image
    cartoon = cv2.bitwise_and(color, color, mask=edges)
```

```python
    # Display the images
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(img)
    axs[0].set_title("Original Image")
    axs[0].axis("off")

    axs[1].imshow(edges, cmap="gray")
    axs[1].set_title("Edge Detection")
    axs[1].axis("off")

    axs[2].imshow(cartoon)
    axs[2].set_title("Cartoonized Image")
    axs[2].axis("off")

    plt.show()
```
